```python
def scrape_(year,startpage,endpage):
    news_url_list=[]
    news_author_list=[]
    news_date_list=[]
    news_desc_list=[]
    for i in range(startpage,endpage):
        url='https://www.ebmnews.com/'+str(year)+'/'+'page/'+str(i)
        print(url)
        driver.get(url)
        time.sleep(2)
        news_url_div=driver.find_elements_by_xpath('//div[@class="listing listing-blog listing-blog-1 clearfix  columns-1"]/article/div/h2/a')
        news_author_div=driver.find_elements_by_xpath('//div[@class="listing listing-blog listing-blog-1 clearfix  columns-1"]/article/div/div/a/i')
        news_date_div=driver.find_elements_by_xpath('//div[@class="listing listing-blog listing-blog-1 clearfix  columns-1"]/article/div/div/span')
        
        if(year==2018):
            news_desc_div=driver.find_elements_by_xpath('//div[@class="listing listing-blog listing-blog-1 clearfix  columns-1"]/article/div/div[2]')
        else:
            news_desc_div=driver.find_elements_by_xpath('//div[@class="listing listing-blog listing-blog-1 clearfix  columns-1"]/article/div/div[3]')
        
        
        for j in range(0,len(news_url_div)):
            news_url_list.append(news_url_div[j].get_attribute('href'))
            
        print('news_author_div',len(news_author_div))
        for j in range(0,len(news_author_div)):
            news_author_list.append(news_author_div[j].text)
            
        for j in range(0,len(news_date_div)):
            news_date_list.append(news_date_div[j].text)
            
        for j in range(0,len(news_desc_div)):
            news_desc_list.append(news_desc_div[j].text)
        
    return news_url_list,news_author_list,news_date_list,news_desc_list                   
```

```python
def  scrape_detail_(url):
    detail=''
    heading=''
    news_category1=''
    news_category2=''
    news_category=''
    print(url)
    driver.get(url)
    time.sleep(2)
    detail=driver.find_element_by_xpath('//div[@class="single-container"]/article/div[2]').text
    heading=driver.find_element_by_xpath('//div[@class="single-container"]/article/div[1]/h1').text
    news_category1=driver.find_element_by_xpath('//div[@class="single-container"]/article/div[1]/div/div/span[1]').text
    try:
        news_category2=driver.find_element_by_xpath('//div[@class="single-container"]/article/div[1]/div/div/span[2]').text
    except:
        news_category2=''
    if(news_category2==''):
        news_category=news_category1
    else:
        news_category=news_category1+','+news_category2
    print('news_category',news_category)
    return heading,news_category
```

```python
news_url_list_18,news_author_list_18,news_date_list_18,news_desc_list_18=scrape_(2018,357,412)
```

```python
heading_18_list=[]
category_18_list=[]  
```

```python
for m in range(0,len(news_url_list_18)):
    heading_18,news_category_18=scrape_detail_(news_url_list_18[m])
    heading_18_list.append(heading_18)
    category_18_list.append(news_category_18)
```
